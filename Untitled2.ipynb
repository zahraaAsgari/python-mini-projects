```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("online_retail.csv", encoding="ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
df.describe()
```

```python
df.dropna(subset=["CustomerID"])
```

```python
df=df[(df["Quantity"]>0) & (df["UnitPrice"]>0)]
```

```python
df["total_price"]=df["Quantity"] * df["UnitPrice"]
df.head()
```

```python
top_countries=df.groupby("Country")["total_price"].sum().sort_values(ascending=False).head(15)
```

```python
df.groupby("Description")["total_price"].sum().sort_values(ascending=False).head(15).plot(kind="bar",figsize=(12,6))

plt.savefig('top 15 customer.jpg',dpi=300,bbox_inches='tight')
```

```python
order_counts=df["Country"].value_counts().head(15)
plt.figure(figsize=(12,5))
sns.barplot(x=order_counts.values, y=order_counts.index)
plt.title("Top 10 Countries by Order")
plt.xlabel("Order")
plt.ylabel("Country")
plt.savefig('CountriesbyOrder.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
top_countries=df.groupby("Country")["total_price"].sum().sort_values(ascending=False).head(15)
plt.figure(figsize=(12,5))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Country")
plt.savefig('Countries by Revenue.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
total_revenue=df["total_price"].sum()
total_orders=df["InvoiceNo"].nunique()
unique_customer=df["CustomerID"].nunique()
avg_order_value=total_revenue/total_orders if total_orders>0 else 0
avg_customer_value=total_revenue/unique_customer if unique_customer>0 else 0

metrics={
   "Total Revenue":total_revenue ,
    "Total Orders":total_orders,
    "Unique Customer":unique_customer,
    "AOV":avg_order_value,
    "ACV":avg_customer_value
}

names=list(metrics.keys())
values=list(metrics.values())


plt.figure(figsize=(8,5))
plt.bar(names, values, color=["red","blue","green","orange","purple"])
plt.title("Business Metrics Overview")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
labels = ["AOV", "ACV", "Orders", "Customers"]
sizes = [avg_order_value, avg_customer_value, total_orders, unique_customer]
colors = ["#6C63FF", "#FF6584", "#3DC1D3", "#FFA534"]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", colors=colors, textprops={"fontsize": 12})
centre_circle = plt.Circle((0, 0), 0.70, fc="white")
plt.gca().add_artist(centre_circle)

plt.title("Key Metrics Overview", fontsize=16, fontweight="bold")
plt.tight_layout()
plt.savefig('overview_key_metrics.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt

# --- Values ---
total_revenue = df["total_price"].sum()
total_orders = df["InvoiceNo"].nunique()
unique_customer = df["CustomerID"].nunique()
avg_order_value = total_revenue / total_orders if total_orders > 0 else 0
avg_customer_value = total_revenue / unique_customer if unique_customer > 0 else 0

metrics = {
    "Total Revenue": total_revenue,
    "Total Orders": total_orders,
    "Unique Customers": unique_customer,
    "Average Order Value": avg_order_value,
    "Avg Customer Value (ACV)": avg_customer_value
}

# --- Plot ---
plt.figure(figsize=(12, 7))
colors = ["#6C63FF", "#FF6584", "#3DC1D3", "#FFA534", "#2ECC71"]

for i, (title, value) in enumerate(metrics.items()):
    plt.barh(title, value, color=colors[i], height=0.5)

plt.title("Business Performance Metrics", fontsize=16, fontweight="bold")
plt.xlabel("Value", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Revenue'] = df['Quantity'] * df['UnitPrice']
reference_date = df['InvoiceDate'].max()
rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'Revenue': 'sum'
})
rfm.columns = ['Recency', 'Frequency', 'Monetary']

best_customers = rfm.sort_values(['Monetary', 'Frequency'], ascending=False).head(10)
print(best_customers.reset_index())
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
sns.histplot(rfm['Recency'], bins=30, kde=True, color='skyblue')
plt.title('Recency Distribution')

plt.subplot(1, 3, 2)
sns.histplot(rfm['Frequency'], bins=30, kde=True, color='orange')
plt.title('Frequency Distribution')

plt.subplot(1, 3, 3)
sns.histplot(rfm['Monetary'], bins=30, kde=True, color='green')
plt.title('Monetary Distribution')

plt.tight_layout()
plt.savefig('RFM.png',dpi=300,bbox_inches='tight')
plt.show()
```
