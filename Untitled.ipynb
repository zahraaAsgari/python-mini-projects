```python
from fredapi import Fred
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import os
from dotenv import load_dotenv

load_dotenv()
fred_key=os.getenv('fred_api_key')
fred= Fred(api_key=fred_key)
series=fred.get_series('RSXFSN')
df=pd.DataFrame(series,columns=['Values'])
df['Date'] = df.index
df=df.reset_index()
df['Date']=pd.to_datetime(df['Date'])

```

```python
print(df.head())
```

```python
df.plot(x='Date', y='Values', figsize=(10,5))
```

```python
average_value=df['Values'].mean()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Values'],label='Data')
plt.axhline(average_value,color='red',linestyle='--',label='Average')
plt.legend()
plt.grid(True)
         
         
```

```python
max_value=df['Values'].max()
plt.figure(figsize=(12,6))
plt.plot(df['Date'],df['Values'],label='Data')
plt.axhline(max_value,color='red',linestyle='--',label='Max')
plt.legend()
plt.grid(True)
```

```python
min_value=df.loc[df['Values'].idxmin()]
plt.figure(figsize=(10,6))
plt.plot(df['Date'],df['Values'],label='Data')
plt.grid(True)
```

```python
df['rolling_12']=df['Values'].rolling(window=12).mean()
df.head()
plt.figure(figsize=(12,6))
plt.bar(df['Date'],df['Values'],label='Original')
plt.plot(df['Date'],df['rolling_12'],label='Rolling 12',linewidth=3,color='red')
plt.legend()
plt.grid(True)
```

```python
max_value=df['Values'].max()
plt.figure(figsize=(12,6))
plt.scatter(df['Date'],df['Values'],label='Data')
plt.axhline(max_value,color='red',linestyle='--',label='Max')
plt.legend()
plt.grid(True)
```

```python
average_value=df['Values'].mean()
plt.figure(figsize=(10,6))
plt.hist(df['Values'],label='Data',bins=20)
plt.axhline(average_value,color='red',linestyle='--',label='Average')
plt.legend()
plt.grid(True)
```
