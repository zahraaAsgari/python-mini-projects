```python
from fredapi import Fred
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

df = pd.read_csv("dubai_properties.csv")
df.head()
# df.shape
```

```python
df = df.dropna(subset=["Latitude", "Longitude"])
```

```python
df['Rent'].mean()
```

```python
df.groupby("Location")["Rent"].mean().sort_values(ascending=False)
```

```python
df["ROI"]=(df["Rent"]*12 / df["Area_in_sqft"])*100
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Area_in_sqft"],df["Rent"])
plt.xlabel("Area")
plt.ylabel("Rent")
plt.title("Area vs Rent")
plt.show()
```

```python
roi_by_are=df.groupby("Location")["ROI"].mean().sort_values(ascending=False)    #کجا بیشترین بازده را دارد
roi_by_are.head(10)
```

```python
roi_by_are.head(10).plot(kind="bar",figsize=(12,6))
```

```python
avg_rent=df.groupby("Location")["Rent"].mean().sort_values(ascending=False)      #گران ترین و ارزان ترین مناطق دبی برای اجاره
avg_rent.head(10)
```

```python
avg_rent.head(10).plot(kind="bar",figsize=(12,6))
```

```python
#چطور تعداد اتاق ها قیمت اجاره را تغییر میدهد
df.groupby("Beds")["Rent"].mean().plot(kind="bar",figsize=(12,6))
```

```python
#ترند اجاره در 12 ماه گذشته چه بوده  است؟رو به رشد یا افت!!!
df["Posted_date"]=pd.to_datetime(df["Posted_date"],errors="coerce")
df=df.dropna(subset=["Posted_date"])
monthly_trend=df.resample("ME",on="Posted_date")["Rent"].mean()
monthly_trend.plot(figsize=(12,6))
```

```python
#select only numeric column
import seaborn as sns
numeric_df = df.select_dtypes(include=['int64','float64'])
plt.figure(figsize=(12,6))
sns.heatmap(numeric_df.corr(),annot=True,cmap='coolwarm',linewidth=0.5)
plt.title('Correlation Heatmap of Dubai Real Estate Market')
plt.savefig('dubai_heatmap.png',dpi=300,bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(24,6))
sns.boxplot(x='Location',y='Rent',data=df)
plt.xticks(rotation=90)
plt.title('Distribution of Rent by Location in Dubai',fontsize=16)
plt.xlabel('Location')
plt.ylabel('Rent (AED)',fontsize=12)
plt.savefig('dubai_rent_boxplot.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

```python
top_locations=df.groupby("Location")["Rent"].median().sort_values(ascending=False).head(10)
df_top=df[df["Location"].isin(top_locations.index)]

plt.figure(figsize=(14,6))
sns.boxplot(x='Location',y='Rent',data=df_top)
plt.xticks(rotation=45,ha='right')
plt.title('Top 10 Locations Rent Distribution in Dubai',fontsize=16)
plt.xlabel('Location')
plt.ylabel('Rent (AED)',fontsize=12)
plt.savefig('dubai_rent_boxplot.jpg',dpi=300,bbox_inches='tight')
plt.show()
```

```python
top_locations=df.groupby("Location")["Rent"].median().sort_values(ascending=False).head(20)
print(top_locations)
```

```python
df["ROI"]=(df["Rent"]*12 / df["Area_in_sqft"])*100
top_roi_locations=df.groupby("Location")["ROI"].mean().sort_values(ascending=False).head(10)
df_top_roi=df[df["Location"].isin(top_roi_locations.index)]
plt.figure(figsize=(12,6))
sns.scatterplot(x="Rent", y="ROI",data=df_top_roi,hue="Location",palette="tab10",alpha=0.6,edgecolor=None)
plt.title("Scatterplot of ROI vs Rent in Dubai Real Estate",fontsize=16)
plt.xlabel("Rent (AED)",fontsize=12)
plt.ylabel("ROI (%)" , fontsize=12)
plt.legend(bbox_to_anchor=(1.05,1),loc="upper left",title="Location")
# plt.tight_layout()
plt.savefig("roi_vs_rent.png",dpi=300,bbox_inches="tight")
plt.show()
```
