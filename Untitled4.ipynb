```python
import pandas as pd
import numpy as np

# فرض: df قبلاً خوانده شده (df = pd.read_csv('your_file.csv'))
print("دیتافریم اولیه - نوع ستون‌ها:")
print(df[['TRANS_VALUE', 'PROCEDURE_AREA', 'ACTUAL_AREA', 'PARKING']].dtypes)
print("\nنمونه مقادیر مشکل‌دار:")
print(df[['TRANS_VALUE', 'PROCEDURE_AREA', 'ACTUAL_AREA', 'PARKING']].head(10))

# لیست ستون‌های عددی که نیاز به تبدیل دارند
numeric_cols = ['TRANS_VALUE', 'PROCEDURE_AREA', 'ACTUAL_AREA', 'PARKING']

# تمیز کردن: تبدیل به string، حذف space و comma، تبدیل به numeric (errors='coerce' برای NaN کردن invalidها)
for col in numeric_cols:
    df[col] = df[col].astype(str).str.replace(r'[\s,]', '', regex=True)  # حذف space و comma
    df[col] = df[col].str.replace(r'M$', '000000', regex=True)  # اگر '4.5M' داشت (نادر)
    df[col] = pd.to_numeric(df[col], errors='coerce')  # تبدیل به float، invalid → NaN

# حذف یا impute ردیف‌های NaN در ستون‌های کلیدی (برای دقت بالا)
df_clean = df.dropna(subset=numeric_cols)  # یا median impute اگر نمی‌خواهید حذف کنید:
# df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

print("\nپس از تمیز کردن - نوع ستون‌ها:")
print(df_clean[['TRANS_VALUE', 'PROCEDURE_AREA', 'ACTUAL_AREA', 'PARKING']].dtypes)
print("\nنمونه پس از تمیز کردن:")
print(df_clean[['TRANS_VALUE', 'PROCEDURE_AREA', 'ACTUAL_AREA', 'PARKING']].head())

# حالا corr بدون مشکل کار می‌کند
corr = df_clean[['TRANS_VALUE', 'PROCEDURE_AREA', 'ACTUAL_AREA', 'PARKING']].corr()
print("\nCorrelation Matrix:")
print(corr)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style='whitegrid', palette='coolwarm')  # شیک برای گزارش مشتری

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='Blues', fmt='.2f', linewidths=0.5)
plt.title('Correlations in Dubai Luxury Properties (2025 Data)')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


sns.histplot(df_clean['TRANS_VALUE'], bins=20, kde=True)  # viz برای درک بصری
plt.title('توزیع ارزش معاملات (AED) - تمرکز روی لوکس')
plt.xlabel('ارزش معامله')
plt.ylabel('تعداد')
plt.show()  # خروجی: ببین چطور skewedه – یعنی فرصت در high-end
```
