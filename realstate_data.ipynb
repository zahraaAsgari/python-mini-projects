```python
important_cols=[
    "Rent",
    "Beds",
    "Baths",
    "Location",
    "Type",
    "Area_in_sqft",
    "City",
    "Posted_date"  
]

df1=df["important_cols"]

df1["Area_in_sqft"].min()
```

```python
bins=[0,600,1200,2500,5000 ,df["Area_in_sqft"].max()]
labels=[
    "micro",
    "standars apt",
    "family apt",
    "luxury residence",
    "ult_luxury/villa"
]
df1["Area_segment"]=pd.cut(df1["Area_in_sqft"],bins=bins,labels=labels)
df1.head(25)

```

```python
df1["Area_segment"].value_counts(normalize=True)*100
```

```python
plt.figure(figsize=(8,4))
sns.countplot(
    y=df1["Area_segment"],
    order=df1["Area_segment"].value_counts().index
)
plt.title("Property size segments in dubai market")
plt.xlabel("number of listing")
plt.ylabel("rent segment")
plt.show()
```

```python
# نمودار نقشه عرضه بازار رو نشون میده
# این نمودار تقاضا رو نشون نمیده
# بیشترین عرضه در family apt وstandard aptهست 
# luxury residence, microحجمشون کمتره اما قیمت گذاری حساس تر و مشتری ها انتخابی تر هستند و هر تحلیل اشتباه خیلی گرون تموم میشه
# کمترین عرضه برای ult_luxuryهست و بازار خاص خودش رو داره و باید جداگانه تحلیل بشه
```

```python
luxury_df=df1[df1["Area_segment"].isin(["luxury residence","ult_luxury/villa"])]
```

```python
luxury_df.head(20)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=luxury_df,x="Area_in_sqft",y="Rent",hue="Area_segment",alpha=0.9)
plt.title("Rent vs Area | Luxury Properties")
plt.xlabel("Area (sqft)")
plt.ylabel("Rent (AED)")
plt.show()
```

```python
import pandas as pd
import numpy as np
df=pd.read_csv("dubai_properties.csv")
crosstab_ab=pd.crosstab(df["Rent"],df["Location"], margins=True, normalize='index')
print(crosstab_ab)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.heatmap(crosstab_ab,annot=True,fmt='.1%',cmap='Blues',cbar_kws={'label':'درصد'})
plt.title("Heatmap|AB Test")
plt.show()
```
